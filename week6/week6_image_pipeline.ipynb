```python
# Cell 1 — Load image from URL
import requests
import numpy as np
from PIL import Image
from io import BytesIO

url = "https://picsum.photos/640"

response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")

img_array = np.array(img)

print("Shape:", img_array.shape)
print("Min pixel:", img_array.min(), "Max pixel:", img_array.max())

img
```

```python
print(response.status_code)
print(response.headers.get("Content-Type"))
```

```python
# Cell 2 — Resize
resized = img.resize((640, 640))
print("Resized shape:", np.array(resized).shape)
resized
```

```python
# Cell 3 — Grayscale
gray = img.convert("L")
gray_array = np.array(gray)
print("Grayscale shape:", gray_array.shape)
print("Channels lost: color info, only luminance remains")
gray
```

```python
# Cell 4 — Crop region of interest
cropped = img_array[50:250, 100:400]
print("Cropped shape:", cropped.shape)
Image.fromarray(cropped)
```

```python
# Cell 5 — Normalize pixel values
normalized = img_array / 255.0
print("Original range: 0 to 255")
print("Normalized range:", normalized.min(), "to", normalized.max())
```
